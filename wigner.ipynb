```python
## Importing bits
import numpy as np
from arc import wigner as w

## NB Needed to use numpy <2 (1.26.4)
```

# Wigner 3j Calculator 

```python
def print_winger3j(js,ms): # both js and ms need to be 3x1 np arays
    return w.Wigner3j(js[0], js[1], js[2], ms[0], ms[1], ms[2])
```

```python
js3 = np.array([1,1/2,3/2])
ms3 = np.array([1,-1/2,-1/2])

result = print_winger3j(js3,ms3)
```

```python
exact_result_2 = (1/(result**2))
print('The surd form is {} / sqrt({})'.format(np.sign(result), exact_result_2))
```
